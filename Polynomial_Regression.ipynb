```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
#Dataset
url="https://raw.githubusercontent.com/apratim777/apratim777/refs/heads/master/Position_Salaries.csv"
df=pd.read_csv(url)
df
```

```python
%matplotlib inline
plt.xlabel("Level")
plt.ylabel("Salary")
plt.scatter(df.Level,df.Salary)
```

```python
y=df.Salary
y
```

```python
x=df.Level
x
```

```python
x=df.iloc[:,1:2].values
x
y=df.iloc[:,2].values
y
```

```python
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.predict([[10]])
```

```python
reg.score(x,y)
```

```python
%matplotlib inline
plt.xlabel("Level")
plt.ylabel("Salary")
plt.scatter(df.Level,df.Salary)
plt.plot(x,reg.predict(x))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
x_poly = poly.fit_transform(x)
x_poly
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_poly,y)
```

```python
reg.score(x_poly,y)
```

```python
reg.predict(x_poly)
```

```python
plt.scatter(x,y,color="red")
plt.plot(x,reg.predict(poly.fit_transform(x)),color="blue")
```
