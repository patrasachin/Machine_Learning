```python
#import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Dataset
url="https://raw.githubusercontent.com/apratim777/apratim777/refs/heads/master/homeprices.csv"
df=pd.read_csv(url)
df
```

```python
#VISUALIZATION

%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.area,df.price)
```

```python
# Assign the 'price' column from the DataFrame as the target variable y

y=df.price
y
```

```python
# Assign the 'area' column from the DataFrame as the input feature x

x=df.area
x
```

```python
#Separate input features (x) by removing the target column 'price'

x=df.drop('price',axis='columns')
x
```

```python
# Drop the 'area' column and keep the remaining column(s) as the target variable y

y=df.drop('area',axis='columns')
x
```

```python
# Create an instance of the Linear Regression model

reg=LinearRegression()
```

```python
# Train (fit) the Linear Regression model using input feature(s) x and target variable y

reg.fit(x,y)
```

```python
#Predict the output

reg.predict([[2600]])
```

```python
#find the coefficient(slope)
reg.coef_
```

```python
#find intercept
reg.intercept_
```
