```python
#Import Librarie

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Dataset
url="https://raw.githubusercontent.com/apratim777/apratim777/refs/heads/master/homeprices1.csv"
df=pd.read_csv(url)
df
```

```python
#Find the median for the missing value in bedrooms

df.bedrooms.median()
```

```python
#Fill the missing value by median

df.bedrooms=df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
# Separate input features (x) by removing the target column 'price'

x = df.drop('price', axis='columns')
x
```

```python
## Drop the feature columns 'area', 'bedrooms', and 'age' to keep only the target column(s) in y

y = df.drop(['area', 'bedrooms', 'age'], axis='columns')
y
```

```python
# Create an instance of the Linear Regression model

reg=LinearRegression()
```

```python
# Train (fit) the Linear Regression model using input feature(s) x and target variable y

reg.fit(x,y)
```

```python
#Predict the output

reg.predict([[3200,4,18]])
```

```python
#find the coefficient(slope)

reg.coef_
```

```python
#find intercept
reg.intercept_
```
